# (i) FIM_1 Q: Test drive Transaction Reduction Variant of Apriori algorithm to generate association rules satisfying minimum support ad confidence

```python
# Import necessary libraries
from itertools import combinations
import pandas as pd
import numpy as np

# Load the IPL matches dataset
df = pd.read_csv("IPL Matches 2008-2020.csv")

# Select the columns of interest
cols = ["id", "date", "team1", "team2", "winner"]
df = df[cols]

# Compute the support of each team
team_support = df["team1"].value_counts() + df["team2"].value_counts()

# Sort the teams by support
team_support = team_support.sort_values(ascending=False)

# Select the top N teams by support
N = 8
top_teams = team_support.head(N).index

# Reduce the dataset by selecting only the matches involving the top teams
df_reduced = df[df["team1"].isin(top_teams) & df["team2"].isin(top_teams)]

# Compute the itemsets and their support
itemsets = {}
for i in range(1, len(top_teams)+1):
    itemsets[i] = {}
    for team in top_teams:
        support = ((df_reduced["team1"] == team) | (df_reduced["team2"] == team)).sum()
        itemsets[i][frozenset([team])] = support
```

Above code bock says, The teams are sorted in descending order by support.The top N teams are selected, where N is set to 8.The dataset is reduced to only include matches involving the top N teams.The itemsets and their support are computed for each size of itemset (from 1 to N).The itemsets are stored in a dictionary itemsets, where the keys are the size of the itemset and the values are dictionaries containing the itemsets and their support.

```python
# Generate candidate itemsets of size k+1 from frequent itemsets of size k
def apriori_gen(Lk, k):
    Ck = set()
    for itemset1 in Lk:
        for itemset2 in Lk:
            if len(itemset1.union(itemset2)) == k+1:
                Ck.add(itemset1.union(itemset2))
    return Ck
```

This function generates candidate itemsets of size 'k+1' from frequent itemsets of size 'k'.
It takes two parameters:
'Lk' is a set of frequent itemsets of size 'k'.
'k' is the size of the frequent itemsets in 'Lk'.
The function returns a set of candidate itemsets of size 'k+1'.

```python
# Compute the frequent itemsets and their support
min_support = 0.05
L = {}
for i in range(1, len(top_teams)+1):
    L[i] = {k: v for k, v in itemsets[i].items() if v/len(df_reduced) >= min_support}
    if len(L[i]) == 0:
        break
    if i < len(top_teams):
        C = apriori_gen(set(L[i].keys()), i)
        for index, row in df_reduced.iterrows():
            for itemset in C:
                if itemset.issubset(set([row["team1"], row["team2"]])):
                    itemsets[i+1][itemset] = itemsets[i+1].get(itemset, 0) + 1
        L[i+1] = {k: v for k, v in itemsets[i+1].items() if v/len(df_reduced) >= min_support}
        if len(L[i+1]) == 0:
            break
```

It starts by setting a minimum support threshold and initializing an empty dictionary L to store the frequent itemsets. It iteratively generates candidate itemsets of size k+1 from frequent itemsets of size k, and filters out those that do not meet the minimum support threshold. If there are no more frequent itemsets of a particular size, the loop terminates.

```python
# Compute the association rules
min_confidence = 0.7
for k in range(2, len(top_teams)+1):
    for itemset in L[k]:
        for i in range(1, k):
            for antecedent in combinations(itemset, i):
                antecedent = frozenset(antecedent)
                consequent = itemset.difference(antecedent)
                if antecedent in L[i]:
                    confidence = L[k][itemset] / L[i][antecedent]
                    if confidence >= min_confidence:
                        print(f"{antecedent} => {consequent}, support: {L[k][itemset]/len(df_reduced)}, confidence: {confidence}")
```

 It loops through all itemsets of length 2 or greater, and generates all possible antecedents and consequents for each itemset. It then calculates the confidence of each association rule and prints those that meet the minimum confidence threshold.

# (ii) FIM_2 Q: Generate Closed Frequent Itemsets by employing LCM algorithm

```python
import csv
from collections import defaultdict

# Load transactions from CSV file
transactions = []
with open('IPL Matches 2008-2020.csv', 'r') as f:
    csv_reader = csv.reader(f)
    next(csv_reader)  # Skip header row
    for row in csv_reader:
        transaction = set([row[6], row[7], row[8], row[9], row[10]])
        transactions.append(transaction)
```

This code reads a dataset file of transactions and generates closed frequent itemsets using the LCM algorithm. The generate_frequent_itemsets function takes the list of transactions and a minimum support count as inputs, and returns a dictionary of closed frequent itemsets and their frequencies.

```python
# Function to generate frequent itemsets
def generate_frequent_itemsets(transactions, min_support):
    # Count the frequency of each item in the transactions
    item_counts = defaultdict(int)
    for transaction in transactions:
        for item in transaction:
            item_counts[item] += 1
    
    # Filter items below the minimum support
    frequent_items = {}
    for item, count in item_counts.items():
        if count >= min_support:
            frequent_items[item] = count
    
    # Convert frequent items to a list of sets for LCM algorithm
    frequent_itemsets = [set([item]) for item in frequent_items.keys()]
    
    # LCM algorithm to generate frequent itemsets
    k = 2
    while True:
        new_itemsets = []
        for i in range(len(frequent_itemsets)):
            for j in range(i+1, len(frequent_itemsets)):
                itemset1 = frequent_itemsets[i]
                itemset2 = frequent_itemsets[j]
                if len(itemset1.union(itemset2)) == k:
                    if all(item in frequent_itemsets for item in itemset1.union(itemset2)):
                        new_itemsets.append(itemset1.union(itemset2))
        if not new_itemsets:
            break
        frequent_itemsets = new_itemsets
        k += 1
    
    # Count the frequency of each frequent itemset in the transactions
    itemset_counts = defaultdict(int)
    for transaction in transactions:
        for itemset in frequent_itemsets:
            if itemset.issubset(transaction):
                itemset_counts[frozenset(itemset)] += 1
    
    # Generate closed frequent itemsets
    closed_frequent_itemsets = {}
    for itemset, count in itemset_counts.items():
        is_closed = True
        for other_itemset, other_count in itemset_counts.items():
            if other_itemset != itemset and itemset.issubset(other_itemset) and count == other_count:
                is_closed = False
                break
        if is_closed:
            closed_frequent_itemsets[itemset] = count
    
    return closed_frequent_itemsets
```

The function first counts the frequency of each item in the transactions and filters out items that occur less frequently than the minimum support count. Then, it converts the frequent items into a list of sets for the LCM algorithm. The LCM algorithm generates frequent itemsets of increasing size until no new itemsets can be generated. The function then counts the frequency of each frequent itemset in the transactions.

Finally, the function generates closed frequent itemsets by checking if each itemset is closed . The closed frequent itemsets and their frequencies are returned as a dictionary.

```python
min_support = 10

closed_frequent_itemsets = generate_frequent_itemsets(transactions, min_support)

# Print the closed frequent itemsets and their frequencies
for itemset, count in closed_frequent_itemsets.items():
    print(itemset, count)
```

The code sets the minimum support count to 10, calls the 'generate_frequent_itemsets' function, and prints the resulting closed frequent itemsets and their frequencies.

# (iii) Predictive Analytics Q: Test drive Back Propagation or ROCK clustering based on the nature of labelled or unlabeled data set.

### Back Propagation

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier

# Load data
data = pd.read_csv('IPL Matches 2008-2020.csv')
data = data.drop(['id', 'date', 'method', 'umpire1', 'umpire2'], axis=1)

# Remove rows with missing values
data.dropna(inplace=True)

# Create label encoder object
le = LabelEncoder()

# Encode categorical columns
data['city'] = le.fit_transform(data['city'])
data['player_of_match'] = le.fit_transform(data['player_of_match'])
data['venue'] = le.fit_transform(data['venue'])
data['team1'] = le.fit_transform(data['team1'])
data['team2'] = le.fit_transform(data['team2'])
data['toss_winner'] = le.fit_transform(data['toss_winner'])
data['toss_decision'] = le.fit_transform(data['toss_decision'])
data['winner'] = le.fit_transform(data['winner'])
data['result'] = le.fit_transform(data['result'])
data['eliminator'] = le.fit_transform(data['eliminator'])
```

This code uses the scikit-learn library to train a Multi-Layer Perceptron (MLP) classifier on data from IPL cricket matches.

First, the data is loaded from a CSV file, certain columns are dropped, and any rows with missing values are removed. Then, categorical columns are encoded using the LabelEncoder object.

```python
# Split data into X and y
X = data.drop(['result'], axis=1)
y = data['result']

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create MLP classifier object
clf = MLPClassifier(hidden_layer_sizes=(10,), activation='logistic', solver='sgd', max_iter=1000, random_state=42)

# Train the model
clf.fit(X_train, y_train)
```

The data is split into training and testing sets using the train_test_split function from scikit-learn.

An MLP classifier object is created with one hidden layer of 10 neurons, the logistic activation function, stochastic gradient descent solver, and a maximum of 1000 iterations.

The model is trained on the training set using the fit method of the MLPClassifier object.

```python
# Evaluate the model
print("Accuracy on training set:", clf.score(X_train, y_train))
print("Accuracy on test set:", clf.score(X_test, y_test))
```

Finally, the accuracy of the model is evaluated on both the training and test sets using the score method of the MLPClassifier object. The accuracy of the model on the training and test sets is printed to the console.

### Pyspark for Transaction Reduction 

```python
from itertools import combinations
from pyspark.sql import SparkSession

# create a spark session
spark = SparkSession.builder.appName("IPL Association Rule Mining").getOrCreate()

# read the dataset
df = spark.read.format("csv").option("header", "true").load("IPL Matches 2008-2020.csv")

# select the columns of interest
cols = ["id", "date", "team1", "team2", "winner"]
df = df.select(cols)

# compute the support of each team
team_support = df.groupBy("team1").count().union(df.groupBy("team2").count())
team_support = team_support.groupBy("team1").sum("count").sort("sum(count)", ascending=False)

# select the top N teams by support
N = 8
top_teams = [row[0] for row in team_support.head(N)]

# reduce the dataset by selecting only the matches involving the top teams
df_reduced = df.filter((df["team1"].isin(top_teams)) & (df["team2"].isin(top_teams)))

# Compute the itemsets and their support
itemsets = {}
for i in range(1, len(top_teams)+1):
    itemsets[i] = {}
    for team in top_teams:
        support = df_reduced.filter((df_reduced["team1"] == team) | (df_reduced["team2"] == team)).count()
        itemsets[i][frozenset([team])] = support

# Generate candidate itemsets of size k+1 from frequent itemsets of size k
def apriori_gen(Lk, k):
    Ck = set()
    for itemset1 in Lk:
        for itemset2 in Lk:
            if len(itemset1.union(itemset2)) == k+1:
                Ck.add(itemset1.union(itemset2))
    return Ck

# Compute the frequent itemsets and their support
min_support = 0.05
L = {}
for i in range(1, len(top_teams)+1):
    L[i] = {k: v for k, v in itemsets[i].items() if v/df_reduced.count() >= min_support}
    if len(L[i]) == 0:
        break
    if i < len(top_teams):
        C = apriori_gen(set(L[i].keys()), i)
        for row in df_reduced.rdd.collect():
            for itemset in C:
                if itemset.issubset(set([row["team1"], row["team2"]])):
                    itemsets[i+1][itemset] = itemsets[i+1].get(itemset, 0) + 1
        L[i+1] = {k: v for k, v in itemsets[i+1].items() if v/df_reduced.count() >= min_support}
        if len(L[i+1]) == 0:
            break

# Compute the association rules
min_confidence = 0.7
for k in range(2, len(top_teams)+1):
    for itemset in L[k]:
        for i in range(1, k):
            for antecedent in combinations(itemset, i):
                antecedent = frozenset(antecedent)
                consequent = itemset.difference(antecedent)
                if antecedent in L[i]:
                    confidence = L[k][itemset] / L[i][antecedent]
                    if confidence >= min_confidence:
                        print(f"{antecedent} => {consequent}, support: {L[k][itemset]/len(df_reduced)}, confidence: {confidence}")
```

The above code performs association rule mining on IPL Matches dataset using the Transaction areduction algorithm. It first selects the top N teams by support, reduces the dataset by selecting only the matches involving the top teams, computes frequent itemsets and their support, and finally computes association rules based on a minimum confidence threshold. The code prints the association rules with their support and confidence.

### Pyspark for the Back Propagation Algorithm

```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import StringIndexer, VectorAssembler
from pyspark.ml.classification import MultilayerPerceptronClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

# create a spark session
spark = SparkSession.builder.appName("IPL Neural Network").getOrCreate()

# read the dataset
df = spark.read.format("csv").option("header", "true").load("IPL Matches 2008-2020.csv")

# drop unnecessary columns
df = df.drop('id', 'date', 'method', 'umpire1', 'umpire2')

# remove rows with missing values
df = df.dropna()

# create label indexer object
label_indexer = StringIndexer(inputCol="result", outputCol="label")

# encode categorical columns
string_cols = ['city', 'player_of_match', 'venue', 'team1', 'team2', 'toss_winner', 'toss_decision', 'winner', 'result', 'eliminator']
indexers = [StringIndexer(inputCol=col, outputCol=col+'_index') for col in string_cols]
pipeline = label_indexer
for indexer in indexers:
    pipeline = pipeline.setStages([indexer])
df = pipeline.fit(df).transform(df)

# assemble feature vector
assembler = VectorAssembler(inputCols=[col+'_index' for col in string_cols if col != 'result'], outputCol='features')
df = assembler.transform(df)

# split data into train and test sets
train_df, test_df = df.randomSplit([0.7, 0.3], seed=42)

# create MLP classifier object
classifier = MultilayerPerceptronClassifier(layers=[len(string_cols)-1, 10, 2], blockSize=128, seed=42)

# train the model
model = classifier.fit(train_df)

# evaluate the model
evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="accuracy")
train_accuracy = evaluator.evaluate(model.transform(train_df))
test_accuracy = evaluator.evaluate(model.transform(test_df))

# print accuracy on train and test sets
print("Accuracy on training set:", train_accuracy)
print("Accuracy on test set:", test_accuracy)
```

The given code is implementing a Multi-Layer Perceptron (MLP) classifier using the scikit-learn library. The dataset used is the IPL Matches dataset from 2008-2020. The data is pre-processed by removing some columns, encoding categorical features using LabelEncoder, and splitting the data into training and test sets. The MLP classifier is then trained on the training data and the accuracy of the model is evaluated on both the training and test sets. The final output prints the accuracy of the model on the training and test sets.
